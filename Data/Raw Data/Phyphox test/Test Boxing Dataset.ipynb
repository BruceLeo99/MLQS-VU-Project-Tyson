# 2. Boxing Data preview
After several studies and considerations, we decided to collect the data at sampling rate of 50Hz. Reasons are:
- Professional boxers take 60-100 ms to throw a jab; 150-200ms to throw a cross and other punches;
- Considering the fact that we are hobby-level boxers, it will take us (and other junior boxers) a slightly longer time to throw punches;
- Thus, we assume that it will take us approximately 250ms to throw a punch. Ideally, a sampling frequency of 250Hz shall be able to record a punching behavior with a high temporal resolution (1ms/datapoint); however, this will result in millions of data points which are computationally expensive, so alternatively, we balance this by reducing the temporal resolution to 10ms/datapoint -> 25Hz
- According to **Nyquist Shannon sampling theorem**, the sampling rate must be twice the analog bandwidth of the incoming signal to record a full-scale event including possible buffers (e.g.: a few seconds to reposition your arm). We take twice of the 25Hz -> 50Hz.
